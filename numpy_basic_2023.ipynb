```python
import numpy as np
np.set_printoptions(linewidth=np.inf)
```

```python
x = np.random.randn(100, 28, 28, 1)
y = x[:, :, :, 0]
print(y.shape)
```

```python
x = np.arange(100).reshape(10, 10)
print('original\n', x)
print()

x[x > 50] = 0
print('modified\n', x)
```

```python
a = np.arange(100).reshape(10, 10)
b = np.arange(100, 200).reshape(10, 10)
c = np.arange(200, 300).reshape(10, 10)

print('a\n', a)
print()

print('b\n', b)
print()

print('c\n', c)
```

```python
concat_axis0 = np.concatenate((a, b, c), axis=0)
print('concat axis0\n', concat_axis0)
print('shape: ', concat_axis0.shape)
print()

concat_axis1 = np.concatenate((a, b, c), axis=1)
print('concat axis1\n', concat_axis1)
print('shape: ', concat_axis1.shape)
```

```python
stack_axis0 = np.stack((a, b, c), axis=0)
print('stack axis0\n', stack_axis0)
print('shape: ', stack_axis0.shape)
print()

stack_axis1 = np.stack((a, b, c), axis=1)
print('stack axis1\n', stack_axis1)
print('shape: ', stack_axis1.shape)
```
